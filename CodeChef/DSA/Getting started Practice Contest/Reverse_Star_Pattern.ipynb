#**Problem**

You're given a number $N$. Print the first $N$ lines of the below-given pattern.
```
    *
   **
  ***
 ****
*****
```

##**Input:**

* First-line will contain the number $N$.


##**Output:**
Print the first $N$ lines of the given pattern.

##**Constraints**

* 1 ≤ N ≤ 200

##**Sample Input 1:**
```
4
```

##**Sample Output 1:**
```
   *
  **
 ***
****
```

##**Sample Input 2:**
```
2
```

##**Sample Output 2:**
```
 *
**
```

##**EXPLANATION:**

* In the first example, we'll print the first 4 lines of the given pattern.

* In the second example, we'll print the first 2 lines of the given pattern

#**Code:**

````python
def star(a,b):

    #return 0 if a is 0
    if a == 0:
        return 0

    #print space multiply by input -1
    print((a-1)*" ",end= "")

    #print stars
    print(b*"*")

    #use recursion function(use function itself inside its function)
    star(a-1,b+1)

#input value for star
a = int(input())

#set initial star value
b = 1

#put value and star value in star function
star(a,b)
````
